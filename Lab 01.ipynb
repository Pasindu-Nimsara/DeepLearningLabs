```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title('3D Plot of Z = X^2 + Y^2')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
import pandas as pd

# Load the dataset from a CSV file
df = pd.read_csv('online_store_customer_data.csv')  # Replace 'path_to_your_file.csv' with the actual path to your CSV file

# Display the first few rows of the dataframe
df.head()
```

```python
features = df[['Age', 'Amount_spent']].dropna()

# Display the first few rows of the selected features
features.head()
```

```python
# Calculate Pearson Correlation
pearson_corr = features.corr()

# Calculate Spearman Correlation
from scipy.stats import spearmanr

spearman_corr, _ = spearmanr(features)
spearman_corr = pd.DataFrame(spearman_corr, index=features.columns, columns=features.columns)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot heatmaps
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Pearson Correlation Heatmap
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', ax=axes[0])
axes[0].set_title('Pearson Correlation')

# Spearman Correlation Heatmap
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', ax=axes[1])
axes[1].set_title('Spearman Correlation')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load the dataset into a DataFrame
df = pd.read_csv('Au_nanoParticle_dataset.csv')

# Display the first few rows of the dataframe
df.head()
```

```python
# Filter the columns
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]

# Display the first 20 samples
new_df.head(20)
```

```python
# Calculate mean, standard deviation, and quartiles
stats = new_df.describe()

# Display the calculated statistics
mean = stats.loc['mean']
std_dev = stats.loc['std']
quartiles = stats.loc[['25%', '50%', '75%']]

print("Mean:\n", mean)
print("\nStandard Deviation:\n", std_dev)
print("\nQuartiles:\n", quartiles)
```

```python
import matplotlib.pyplot as plt

# Plot histograms in a 1x4 layout
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

# Plot each histogram
for i, column in enumerate(new_df.columns):
    axes[i].hist(new_df[column], bins=20, color='skyblue', edgecolor='black')
    axes[i].set_title(f'{column} Histogram')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Create pairplot
sns.pairplot(new_df)
plt.show()
```

```python
# Create PairGrid and customize plots
g = sns.PairGrid(new_df)

# Plot bivariate histograms on the upper triangle
g.map_upper(sns.histplot, bins=20)

# Plot histograms with KDE on the diagonal
g.map_diag(sns.histplot, kde=True)

# Plot bivariate KDE on the lower triangle
g.map_lower(sns.kdeplot)

# Display the plot
plt.show()
```
